```python
import torch
from torch.autograd import Variable
import torch.nn.functional as F #변화
# x가 Variable일 때, x.data는 그 값을 갖는 Tensor이며 
# x.grad는 어떤 스칼라 값에 대해 x의 변화도를 갖는 또 다른 Variable 입니다.

# Tensor,Variable 비슷. 차이; autograd가 자동으로 변화도를 계산할 수 있다
#logistic regression
x_data = Variable(torch.Tensor([[1.0],[2.0],[3.0],[4.0]]))
y_data = Variable(torch.Tensor([[0.],[0.],[1.],[1.]])) #0과 1 중 하나

# class 함수 이용해서 linear 모델 만들기
class Model(torch.nn.Module):
  def __init__(self):
    super(Model, self).__init__() #처음 만든다
    self.linear = torch.nn.Linear(1,1) #1 input->1 output
  def forward(self, x):  #x가 인풋이다
    y_pred = F.sigmoid(self.linear(x)) #변화
    return y_pred
 #모델 실행   
model=Model()

# loss, optimizer 만들기
criterion = torch.nn.BCELoss(size_average=True) #내장함수. loss #변화
optimizer = torch.optim.SGD(model.parameters(),lr=0.01)

#Training loop
for epoch in range(1000):
  y_pred = model(x_data) #forward pass

  loss = criterion(y_pred,y_data)
  print(epoch,loss.data)

  optimizer.zero_grad()
  loss.backward() #backward
  optimizer.step()

#After training
hour_var = Variable(torch.Tensor([[1.0]])) #변화
print("predict (after training)", 4, model.forward(hour_var).data[0][0])
```
