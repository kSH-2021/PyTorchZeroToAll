```python
import torch
from torch.autograd import Variable

x_data = Variable(torch.Tensor([[1.0],[2.0],[3.0]]))
y_data = Variable(torch.Tensor([[2.0],[4.0],[6.0]]))

# class 함수 이용해서 linear 모델 만들기
class Model(torch.nn.Module):
  def __init__(self):
    super(Model, self).__init__() #처음 만든다
    self.linear = torch.nn.Linear(1,1) #1 input->1 output
  def forward(self, x):  #x가 인풋이다
    y_pred = self.linear(x) 
    return y_pred
 #모델 실행   
model=Model()

# loss, optimizer 만들기
criterion = torch.nn.MSELoss(size_average=False) #내장함수. loss
optimizer = torch.optim.SGD(model.parameters(),lr=0.01)

#Training loop
for epoch in range(500):
  y_pred = model(x_data) #forward pass

  loss = criterion(y_pred,y_data)
  print(epoch,loss.data)

  optimizer.zero_grad()
  loss.backward() #backward
  optimizer.step()

#After training
hour_var = Variable(torch.Tensor([[4.0]]))
print("predict (after training)", 4, model.forward(hour_var).data[0][0])
```
